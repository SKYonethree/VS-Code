# Triangle Assignment

Sum of two sides > third side

```python
class Triangle:

    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def isTriangle(self):
        a = self.a
        b = self.b 
        c = self.c
        if (a+b>c) and (b+c>a) and (a+c>b):
            t = True
        else:
            t = False
        return t
    
    def perimeter(self):
        if self.isTriangle():
            p = self.a + self.b + self.c
        else:
            print(f'The given sides {self.a}, {self.b}, {self.c} DO NOT FORM A TRIANGLE')
            p = None
        return p 
    
    def area(self):
        if self.isTriangle():
            s = (self.a + self.b + self.c)/2
            a = (s*(s - self.a)*(s - self.b)*(s - self.c))**(1/2)
        else:
            print(f'The given sides {self.a}, {self.b}, {self.c} DO NOT FORM A TRIANGLE')
            a = None
        return a 
```

### Implementing polymorphism in classes

```python
class India:

    def capital(self):
        print("New Delhi is the Capital of India")

    def language(self):
        print("Hindi is widely spoken in India")
```

```python
class USA:

    def capital(self):
        print("Washington D.C. is the Capital of USA")

    def language(self):
        print("English is widely spoken in USA")
```

```python
class France:
    def capital(self):
        print("Paris is the capital of France")

    def language(self):
        print("French is widely spoken in France")
```

```python
c1 = India()
c2 = USA()
c3 = France()
```

```python
c1.language()
```

```python
c2.language()
```

```python
c3.language()
```

```python
for i in [c1, c2, c3]:
    print(type(i))
    i.capital()
    i.language()
    print('\n===========================\n')
```

### Polymorphism to calculate area,perimeter

```python
class Rectangle:

    def __init__ (self, width, height):
        self.width = width
        self.height = height

    def perimeter(self):
        return 2*(self.width + self.height)
    
    def area(self):
        return self.width*self.height
```

```python
from math import pi
pi
```

```python
class Circle:

    def __init__(self, radius):
        self.radius = radius

    def perimeter(self):
        return 2*pi*self.radius
    
    def area(self):
        return pi*(self.radius**2)
```

```python
s1 = Triangle(13, 14, 15)
s2 = Circle(14)
s3 = Rectangle(10, 15)
s4 = Triangle(1, 2, 3)
s5 = Circle(10)
s6 = Rectangle(20, 12)
```

```python
shapes = [s1, s2, s3, s4, s5, s6]
```

```python
shapes
```

```python
for i in shapes:
    print(type(i))
    print(f'Perimeter : {i.perimeter()}')
    print(f'Area : {i.area()}')
    print('\n=====================================\n')
```
