### Two different functions
1. Calculate Hypotenuse for given sided of right angled triangle
2. Area of the triangle for given base and height

```python
def hypotenuse(a, b):
    print(f'Hypotenuse function started')
    time.sleep(3)
    h = (a**2 + b**2)**(1/2)
    print(f'Hypotenuse for given sides {a} and {b} is {h:.4f}')
```

```python
def area(base, height):
    print(f'Area function started')
    time.sleep(3)
    a = (1/2)*base*height
    print(f'Area of the given triangle is {a:.4f}')
```

```python
%%time
hypotenuse(3, 4)
```

```python
%%time
area(20, 10)
```

### Single threaded Execution

```python
%%time
hypotenuse(12, 13) # 3 seconds
area(30, 14) # 3 seconds
```

### Multi-Threaded Execution

```python
from threading import Thread
```

```python
# Create the threads
th1 = Thread(target=hypotenuse, args=(13, 14))
th2 = Thread(target=area, args =(30, 20))

# execute the threads
th1.start()
th2.start()

# Wait for all threads to finish
th1.join()
th2.join()
```

### Assign thread for one function but multiple values

```python
def square(n):
    time.sleep(2)
    s = n**2
    print(f'Square of given number {n} is {s}\n')
```

```python
%%time
square(16)
```

### Single Threading

```python
a = [1, 2, 11, 12, 15]
```

```python
%%time
for i in a:
    square(i)
```

### Multithreading assigning each thread a different number

```python
%%time
# Create the thread
threads = []
for i in a:
    th = Thread(target=square, args=(i,))
    th.start()
    threads.append(th)

# Wait for all threads to finish
for th in threads:
    th.join()
```

```python
import numpy as np
nums = np.arange(1, 100, 1)
nums
```

```python
%%time
threads  = []
# Create and start the threads
for i in nums:
    th = Thread(target=square, args=(i,))
    th.start()
    threads.append(th)

# Wait for all threads to finsh
for th in threads:
    th.join()
```

```python
from urllib.request import urlretrieve
```

```python
def download_file(url):
    filename = url.split('/')[-1]
    print(f'Downloading : {filename} ...')
    urlretrieve(url, filename)
    print(f'{filename} Downloaded')
```

```python
urls[0]
```

```python
download_file(urls[0])
```

```python
%%time
for i in urls:
    download_file(i)
```

### Multi-threading

```python
%%time
# Create all threads and start
threads = []
for i in urls:
    th = Thread(target=download_file, args =(i,))
    th.start()
    threads.append(th)
# Wait for all threads to finish
for th in threads:
    th.join()
```
